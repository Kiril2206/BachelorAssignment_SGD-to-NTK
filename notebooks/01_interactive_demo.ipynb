```python
import matplotlib.pyplot as plt
import json
from matplotlib.ticker import MaxNLocator

# 1. Load Data
filepath = "C:\\Users\\kiril\\Thesis\\results\\logs\\NTK-Wine-Regression_ntk_3_integrated_metrics.json" 
with open(filepath, 'r') as f:
    data = json.load(f)
    loss_history = data['loss']
    switch_epoch = data.get('switch_epoch')  # <--- Read the saved epoch

# 2. Setup X-Axis
epochs_x = range(len(loss_history)) 

# 3. Plot
plt.figure(figsize=(10, 6))
plt.plot(epochs_x, loss_history, marker='o', label='Training Loss')

# Plot the Switch Line dynamically
if switch_epoch is not None:
    plt.axvline(x=switch_epoch, color='red', linestyle='--', label=f'Switch (Epoch {switch_epoch})')

# Highlight the final point (NTK Result)
plt.plot(epochs_x[-1], loss_history[-1], 'r*', markersize=15, label='NTK Result')

plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))
plt.xlim(0, len(loss_history) - 1) 
plt.title("Training Loss (SGD -> NTK)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
